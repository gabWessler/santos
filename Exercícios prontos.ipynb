```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
caminho_arquivo=r'C:\Users\gabriel_w_costa\Downloads\Frutas.xlsx'
```

```python
df_frutas = pd.read_excel(caminho_arquivo)
```

```python
df_frutas
```

```python
df_frutas = df_frutas['Fruta     Cor        Gosto      Formato    Casca'].str.split(expand=True)
```

```python
df_frutas.columns = ['Frutas', 'Cor', 'Sabor', 'Formato', 'Textura']
```

```python
df_frutas
```

```python
y = df_frutas['Frutas']
```

```python
X = df_frutas[['Cor', 'Sabor', 'Formato', 'Textura']]
```

```python
mapeamento_cor = {'Verde': 0, 'Amarelo': 1, 'Marrom': 2}
mapeamento_sabor = {'Azedo': 0, 'Doce': 1}
mapeamento_formato = {'Oval': 0, 'Curvado': 1, 'Alongado':2}
mapeamento_textura = {'Liso': 0, 'Peludo': 1}
```

```python
X['Cor'] = X['Cor'].map(mapeamento_cor)
X['Sabor'] = X['Sabor'].map(mapeamento_sabor)
X['Formato'] = X['Formato'].map(mapeamento_formato)
X['Textura'] = X['Textura'].map(mapeamento_textura)
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=37)
```

```python
modelo = DecisionTreeClassifier()
```

```python
modelo._fit(X_train, y_train)
```

```python
previsoes = modelo.predict(X_test)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
dias = np.arange(1, 31)
vendas = np.random.randint(50, 150, size=30)
```

```python
df_vendas = pd.DataFrame({'Dia': dias, 'Vendas': vendas})
```

```python
df_vendas
```

```python
plt.figure(figsize=(10,6))
plt.plot(df_vendas['Dia'], df_vendas['Vendas'],marker='o', linestyle='-', color='r')
plt.title('Acompanhamento de Vendas Diárias', fontsize=16)
plt.xlabel('Dia do Mês', fontsize=14)
plt.ylabel('Vendas Diárias', fontsize=14)
plt.grid(False)
plt.show()
```

```python
categorias = ['Eletrônicos', 'Roupas', 'Alimentos', 'Móveis', 'Livros']
vendas_totais = [1200, 850, 950, 500, 750]
```
