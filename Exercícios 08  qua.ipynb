```python
import pandas as pd
import numpy as np
import openpyxl
```

```python
num_rows = 50
```

```python
data = {
    'Produto': [f'Produto_{i}' for i in range(1, num_rows + 1)],
    'Categoria': np.random.choice(['Televisão', 'Computador', 'Celular', 'Tablet', 'Impressora'], size=num_rows),
    'Preço': np.random.uniform(100.0, 2000.0, size=num_rows).round(2),
    'Quantidade_Estoque': np.random.randint(1, 50, size=num_rows)}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df['Multiplicação'] = df['Preço'] * df['Quantidade_Estoque'] #2
```

```python
def multiplicar(x):
    return x['Preço'] * x['Quantidade_Estoque'] #3
```

```python
df['Resultado'] = df.apply(multiplicar, axis = 1)
```

```python
df
```
